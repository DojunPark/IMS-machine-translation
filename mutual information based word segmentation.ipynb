```python
with open('../../IMS-machine-translation/data/chinese_original.txt', 'r', encoding='utf-8') as f:
    chinese = f.read()
    
with open('../../IMS-machine-translation/data/korean_original.txt', 'r', encoding='utf-8') as f:
    korean = f.read()
```

# word segmentation

```python
from wordseg import WordSegment

ws_chinese = WordSegment(chinese, max_word_len=2, min_aggregation=1, min_entropy=0.5)
ws_korean = WordSegment(korean, max_word_len=2, min_aggregation=1, min_entropy=0.5)
```

```python
chinese_segmented = ws_chinese.segSentence(chinese)
korean_segmented = ws_korean.segSentence(korean)
```

# Sample sentences segmented

```python
chinese_txt = ''
for i, v in enumerate(chinese_segmented):
    chinese_txt += v + ' '
    if i > 300:
        break
print(chinese_txt)
```

```python
korean_txt = ''
for i, v in enumerate(korean_segmented):
    korean_txt += v + ' '
    if i > 300:
        break
print(korean_txt)
```

# number of types

```python
chinese_set = list(set(chinese_segmented))
print(len(chinese_set))
```

```python
korean_set = list(set(korean_segmented))
print(len(korean_set))
```
